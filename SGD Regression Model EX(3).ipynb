```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_absolute_error , mean_squared_error , median_absolute_error
```

```python
# load_data

path = "G:\\Data Field\\D..S..Course\\data.sc.Asem\\Data\\Data\\2.1 Linear Regression\\houses.csv"
data = pd.read_csv(path)
data
```

```python
# Split_data into input , output

X = data.iloc[ : , : -1]

y = data.iloc[ : , -1]
```

```python
X
```

```python
y
```

```python
# Cleansing_data

Cleansing_data = SimpleImputer(missing_values= np.nan , strategy="mean")

X = Cleansing_data..fit_transform(X)

X

```

```python
# Rescalling data 

Rescalling_data = StandardScaler()
X = Rescalling_data.fit_transform(X)
X
```

```python
# Spliting data into Training and Testing data 

X_train , X_test , y_train , y_test = train_test_split( X , y , test_size = 0.33 , random_state =42 , shuffle = True )
```

```python
# Apply SGD_Model and fit Model

SGD_Reg = SGDRegressor(alpha=1.5 , random_state = 42, loss ="squared_loss" , penalty = "l2" , max_iter=1000 , tol = 1e-3  )

SGD_Reg.fit(X_train , y_train)
```

```python
# Details

#Cofficient Value
Model_Coef = SGD_Reg.coef_
#intercept Value
Model_intercept = SGD_Reg.intercept_

# Trianing Score
Traning_Score = SGD_Reg.score(X_train , y_train)
#Testing Score 
Testing_Score = SGD_Reg.score(X_test , y_test)
#Prediction
y_pred = SGD_Reg.predict(X_test)

#Model traning_Score 

Model_Traning_Score = SGD_Reg.score(X_train , y_train)

#Model Testing_Score

Model_Testing_Score = SGD_Reg.score(X_test , y_test)

# Get Metrics of Model

# Mean_Absolute_Error
MAE = mean_absolute_error(y_test, y_pred , multioutput="uniform_average")
# Mean_squared_Error
MSE = mean_squared_error(y_test, y_pred , multioutput="uniform_average")
# Median_Absolute_Error
Md_A_E = median_absolute_error(y_test, y_pred )

```
